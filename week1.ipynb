````python
# ë¬¸ì œ 1ì˜ ë‹µì•ˆì„ ì‘ì„±í•´ì£¼ì„¸ìš”
import requests
from pprint import pprint

portfolio = {
    "KRW-BTC": 0.1,
    "KRW-ETH": 3,
    "KRW-XRP": 1000,
    # "KRW-SOL": 10
}

def get_has_market_list(portfolio):
  value = []
  for market in portfolio:
    value.append(market)
  return value

def get_current_prices_api():
  url = "https://api.upbit.com/v1/ticker"
  headers = {"accept": "application/json"}
  has_market = get_has_market_list(portfolio)
  markets_param = ",".join(has_market)
  params = {"markets": markets_param}

  response = requests.get(url, headers=headers, params=params)
  result = response.json()

  market_price_dict = {}
  for data in result:
    market_price_dict[data["market"]] = data["trade_price"]
  return market_price_dict

def analyze_portfolio():
    prices = get_current_prices_api()
    portfolio_analysis = []
    total_value = 0

    for market in prices:
        price = prices[market]
        count = portfolio[market]
        value = count * price
        portfolio_analysis.append({
            "name": market,
            "count": count,
            "price": price,
            "value": value
        })
        total_value += value

    print("-" * 80)
    print(f"{'ì½”ì¸':>20} {'ìˆ˜ëŸ‰':>12} {'í˜„ì¬ê°€':>12} {'ê°€ì¹˜':>12} {'ë¹„ì¤‘':>12}")
    print("-" * 80)

    for market in portfolio_analysis:
        coin = market["name"]
        count = market["count"]
        price = int(market["price"])
        value = int(market["value"])
        percent = value / total_value * 100
        print(f"{coin:>20} {count:>12} {price:>12,} {value:>12,} {percent:>12.2f}%")

analyze_portfolio()
````

````python
# ë¬¸ì œ 3ì˜ ë‹µì•ˆì„ ì‘ì„±í•´ì£¼ì„¸ìš”
import requests
from datetime import datetime, timedelta, timezone

KST = timezone(timedelta(hours=9))

def get_investment_day(days_ago):
  now = datetime.now(KST)
  target_date = now - timedelta(days=days_ago)
  return target_date.strftime("%Y-%m-%d")

def get_historical_data_api(market_name, day): 
  url = "https://api.upbit.com/v1/candles/days"
  headers = {"accept": "application/json"}
  params = {
      "market": market_name,
      "count": day
  }
  response = requests.get(url, headers=headers, params=params)
  result = response.json()
  trade_price = list(reversed(result))[0]["trade_price"]
  return int(trade_price)

def get_single_price_api(market_name):
  url = "https://api.upbit.com/v1/ticker"
  headers = {"accept": "application/json"}
  params = {"markets": market_name}
  response = requests.get(url, headers=headers, params=params)
  result = response.json()
  trade_price = result[0]["trade_price"]
  return int(trade_price)

def calculate_investment_return(market_name, days_ago, investment):
  prev_price = get_historical_data_api(market_name, days_ago)
  quantity = round(investment / prev_price, 8)
  current_price = get_single_price_api(market_name)
  current_value = int(quantity * current_price)
  profit = current_value - investment
  return_rate = round((profit / investment) * 100, 1)

  investment_date = get_investment_day(days_ago)


# íˆ¬ì ì¼ì, íˆ¬ì ê¸ˆì•¡, íˆ¬ì ì‹œì  ê°€ê²©, êµ¬ë§¤ ìˆ˜ëŸ‰, í˜„ì¬ ê°€ê²©, í˜„ì¬ ê°€ì¹˜, ì†ìµ, ìˆ˜ìµë¥ ì„ ìˆœì„œëŒ€ë¡œ ì¶œë ¥í•œë‹¤.
  print(f"""
  íˆ¬ì ì¼ì : {investment_date}
  íˆ¬ìê¸ˆì•¡ : {investment:,}ì›
  íˆ¬ì ì‹œì  ê°€ê²© : {prev_price:,}ì›
  êµ¬ë§¤ ìˆ˜ëŸ‰ : {quantity}ì½”ì¸
  í˜„ì¬ ê°€ê²© : {current_price:,}ì›
  í˜„ì¬ ê°€ì¹˜ : {current_value:,}ì›
  ì†ìµ : {profit:,}ì›
  ìˆ˜ìµë¥  : {return_rate}%
  """)
  

calculate_investment_return("KRW-BTC", 30, 1000000000)
````
